```python
import pandas as pd
from IPython.display import display
```

```python
anime_data = pd.read_csv("data/anime.csv")
anime_data
```

```python
# Bersihin Genres dari karakter [] dan '
anime_data['Genres'] = anime_data['Genres'].str.strip("[]").str.replace("'", "").str.split(',').str[0].str.strip()

# Demographics dan Themes tetep sama
anime_data['Demographics'] = anime_data['Demographics'].str.strip("[]").str.replace("'", "")
anime_data['Themes'] = anime_data['Themes'].str.strip("[]").str.replace("'", "")

# Cek hasilnya
print(anime_data['Genres'].head())
print(anime_data['Demographics'].head())
print(anime_data['Themes'].head())
```

```python
anime_data.to_csv("data/clean_anime.csv", index=False)
```
